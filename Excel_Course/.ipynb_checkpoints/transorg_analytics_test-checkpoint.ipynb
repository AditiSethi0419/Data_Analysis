```python
#3
def remove_duplicates(duplicate):
    final_list=[]
    for num in duplicate:
        if num not in final_list:
            final_list.append(num)
    return final_list
duplicate=[1,2,2,1,3,4]
print(remove_duplicates(duplicate))
```
